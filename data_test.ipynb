### Section 3: Downloading data

```python
#getting the list of asx companies from the reference spreadsheet
code_list = list(df_asx['ASX_code_amended'])
```

```python
#creating a date range for the data extraction
start_date = datetime(2022,7,1)
end_date = datetime(2023,6,19)

data = yf.download('QAN.AX', start=start_date, end=end_date)
data
```

```python
#creating empty dataframe to store information
new_df = pd.DataFrame(columns = ['Date', 'Ticker', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'])
```

```python
#looping through the codes in the list...
for code in code_list:
  data = yf.download(code, start=start_date, end=end_date)
  data.insert(0,'Ticker', code)
  new_df = pd.concat([new_df, data])
```

```python
new_df.to_csv('asx tickers data.csv')
```
