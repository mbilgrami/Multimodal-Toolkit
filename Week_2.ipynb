### **Random Forests**

The theory behind random forests is that they generate multiple decision trees based on a random subset of the data (which gives the data a bit of variability).

For classification problems, the class with the 'maximum' trees pointing to it is selected.

For regression problems, the 'mean' of the trees' output is taken.

The example below uses 50 trees

```python
#Training Random Forest
mRF = H2ORandomForestEstimator()
mRF.train(iris.columns[:4], "class", train)
```

```python
#getting results for individual trees. Note how the error comes down after each tree
mRF
```

```python
#getting individual predictions using Random Forest
p = mRF.predict(test)
p
```

```python
#getting overall prediction accuracy
perf = mRF.model_performance(test)

perf
```

### **Gradient Boosting Models (GBMs)**

Gradient Boosting Models are models that work by boosting certain rows for each tree giving each row in your training data a different weight for each tree.

#### **How do GBMs work conceptually?**

GBMs look at the first tree and sees which of the training rows that were easy to learn, which are the ones that were difficult to learn. And the ones that were difficult to learn, it gives a bonus.

If we think back to an earlier video, we looked at a decision tree for learning the high risk dataset. And the setosa were really easy to learn, the other two classes were a bit more complicated, they overlap. So, it would be boosting those other two classes for the second tree.

And it goes ahead and makes the second tree and inevitably, it's going to be a slightly different tree. And this is good, we want slightly different trees. And then, the process repeats.

We look at that second tree and say, "Which were the hardest training rows to learn? Let's boost them again." And you can see eventually, if you have one really difficult training row, all the trees keep getting wrong. Eventually, it's going to get boosted and boosted until it gets a dedicated tree. That may not happen especially not with large datasets, but conceptually, that's what's happening. And that's both the good side and the downside.

#### **Upside of GBMs**
The good side, GBMs can deal really well with non-linearity in your data, awkward data. It can always find the decision boundary however complex it looks.

#### **Downside of GBMs**
The downside, is it's also going to try and learn all the noise in your data. So, when using GBMs, they're very powerful technique but do take extra care to watch out for overfitting.

### **Splitting into training, validation and test sets**


```python
h2o.init()
```

```python
#importing people dataset
import pandas as pd
people_df = pd.read_csv("EmployeeData.csv")
```

```python
people = h2o.H2OFrame(people_df)
```

```python
train, valid, test = people.split_frame([0.8, 0.1], ["people_train", "people_valid", "people_test"], seed = 42)
```

```python
print("%d/%d/%d" % (train.nrows, valid.nrows, test.nrows))
```

```python
from h2o.estimators import H2OGradientBoostingEstimator
```

```python
people.columns
```

```python
#setting up columns
y = "salary"
ignore_fields = [y, "id"]
x = [j for j in people.columns if j not in ignore_fields]
```

Training the model below.

Note that we are also now including an additional field called validation_frame where we are specifying the validation dataset

```python
ml = H2OGradientBoostingEstimator(model_id = "defaults", max_depth = 10, ntrees = 500)
ml.train(x, y, train, validation_frame = valid)
```

```python
#gettingt the mean absolute error of the training and validation datasets
print(ml.mae(train = True))
print(ml.mae(valid = True))
```

```python
#getting the mae of the test data
perf = ml.model_performance(test)
```

```python
perf.mae()
```

#### **Overfitting a GBM**

We are going to create another model and overfit it by increasing the max depth. Although the training MAE is super low compared to our first model, the test and validation MAE is much higher. This indicates an overfit.

```python
#creating another GBM with 1000 trees
m2 = H2OGradientBoostingEstimator(model_id = "overfit", ntrees = 10000, max_depth = 10)
m2.train(x,y, train, validation_frame = valid)
```

```python
print(ml.mae(train = True), m2.mae(train=True))
print(ml.mae(valid = True), m2.mae(valid=True))
print(perf.mae(), m2.model_performance(test).mae())
```

Despite a far lower training MAE, the m2 model has an equivalent validation MAE and a higher testing MAE. This model is overfit.
